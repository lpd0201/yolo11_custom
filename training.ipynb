```python
from ultralytics import YOLO
model = YOLO("/kaggle/working/yolo11_custom/runs/detect/train/weights/best.pt")
```

```python
rs = model.val(data='VisDrone.yaml', imgsz=640, batch=1, device=0)
speed = rs.speed
inf_time = speed['inference']
fps = 1000.0 / inf_time
print(f"Preprocess: {speed['preprocess']:.2f} ms/img")
print(f"Inference: {speed['inference']:.2f} ms/img")
print(f"Postprocess: {speed['postprocess']:.2f} ms/img")
print(f"FPS, batch=1; {fps:.2f}")   
```
